```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/messy_employee_dataset_v3.csv')
def clean_data(df):
  df_copy = df.copy()
  print(df_copy.head(50))
  print(df_copy.info())



  target = 'salary'
  df_copy["age"] = pd.to_numeric(df_copy["age"], errors="coerce") #changin any element in age column that is not a number
  df_copy[target] = pd.to_numeric(df_copy[target], errors="coerce") #changin any element in age column that is not a number


  df_copy = df_copy.replace(["unknown",'missing'], np.nan)
  missing = df_copy.isnull().sum()
  print(missing , '\n the age column is most column that has missing values')

  df_copy["age"] = df_copy["age"].fillna(df_copy["age"].mean())#fill the missing values with mean of the values
  df_copy[target] = df_copy[target].fillna(df_copy[target].median())#fill the missing values with median of the values
  df_copy["years_experience"] = df_copy["years_experience"].fillna(df_copy["years_experience"].mean())#fill the missing values with mean of the values
  df_copy["weekly_hours"] = df_copy["weekly_hours"].fillna(df_copy["weekly_hours"].mean())#fill the missing values with mean of the values
  df_copy["performance_score"] = df_copy["performance_score"].fillna(df_copy["performance_score"].mean())#fill the missing values with mean of the values
  df_copy["department"] = df_copy["department"].fillna(df_copy["department"].mode()[0])#fill the missing values with mode of the values
  df_copy["education_level"] = df_copy["education_level"].fillna(df_copy["education_level"].mode()[0])#fill the missing values with mode of the values
  df_copy["employment_type"] = df_copy["employment_type"].fillna(df_copy["employment_type"].mode()[0])#fill the missing values with mode of the values
  df_copy['projects_handled'] = df_copy['projects_handled'].fillna(df_copy['projects_handled'].mean())#fill the missing values with mean of the values
  df_copy['bonus'] = df_copy['bonus'].fillna(df_copy['bonus'].median())#fill the missing values with mean of the value
  df_copy["region"] = df_copy["region"].fillna(df_copy["region"].mode()[0])#fill the missing values with mode of the values


  canonical_map = {"bachelor": "Bachelor",'sales':'Sales'}
  df["education_level"] = df["education_level"].map(canonical_map).fillna(df["education_level"])
  df["department"] = df["department"].map(canonical_map).fillna(df["department"])


  mask_full = df_copy.duplicated().sum()
  print(mask_full)
  df_copy =  df_copy.drop_duplicates()
  mask_full = df_copy.duplicated().sum()



  q1, q3 = df_copy[target].quantile(0.25), df_copy[target].quantile(0.75)
  iqr = q3 -q1
  lower  = q1 - 1.5 * iqr
  upper = q3 + 1.5 * iqr
  mask_iqr = df_copy[(df_copy[target] < lower) | (df_copy[target] > upper)]
  print(mask_iqr)

  upper_99 = df_copy[target].quantile(0.99)
  df_copy[target] = df_copy[target].clip(upper=upper_99)

  #checking
  for i in df_copy[target]:
    if i  < 0 :
      print('there is value under 0 in target column')


  null_values = df_copy[['age', target]].isnull()

  if True not in null_values :
      print('there is no null values in the target column and the age column')


  df_shape = df.shape
  df_copy_shape = df_copy.shape

  for i in df_shape :
    pass
  for j in df_copy_shape:
    pass
    if j == i :
      print(' the number of columns are the same')




  return df_copy

df =clean_data(df)

```
